### Ahora, veamos como podemos crear una columna artificial en nuestro dataset:

## One-Hot Encoding
### Mujer: (1,0)
### Hombre (0,1)

## Binary Encoding
### Mujer: (1)
### Hombre (0)

```python
df['Americas'] = 0 #La vamos a rellenar solamente con '0' en un comienzo 
df.head()
```

```python
# Relllenamos los valores de la columna
df['Africa'] = df['region'].replace(['Africa','Asia','Europa','Americas','Oceania'],[1,0,0,0,0])
```

# Me quede atrasado

```python
df['region'].unique()
```

```python
for reg in df['region'].unique():
    df[reg] = np.where(df['region'] == reg , 1 , 0)
df.sample(10)
# Lo mismo anterior pero dinamico
```
